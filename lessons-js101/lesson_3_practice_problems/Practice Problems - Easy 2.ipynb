## Question 1
Given a string, return a new string that replaces every occurrence of the word "important" with "urgent":

````javascript
let advice = "Few things in life are as important as" 
advice += " house training your pet dinosaur.";
````

#### Solution 1

````javascript
let newAdvice = advice.split(" ")
      .map((word) => word === "important" ? "urgent" : word)
      .join(" ")
console.log(newAdvice)
````

#### Solution 2
Using [string.replaceAll()](https://developer.mozilla.org/en-US/docs/Web/JavaScript/Reference/Global_Objects/String/replaceAll). This replaces all the occurrences of the word "important". 

SIDE NOTE: For some reason it doesn't work all the time, check your javascript version.

````javascript
let newAdvice2 = advice.replaceAll("important", "urgent")
````

#### Solution 3
Using `replace()`. This will replace ONLY the first occurrence of the word "important".

````javascript
advice.replace('important', 'urgent');
````

## Question 2
The `Array.prototype.reverse` method reverses the order of elements in an array, and `Array.prototype.sort` can rearrange the elements in a variety of ways, including descending. 

Both of these methods mutate the original array as shown below. Write two distinct ways of reversing the array without mutating the original array. Use reverse for the first solution, and `sort` for the second.

````javascript
let numbers = [1, 2, 3, 4, 5];
````

````javascript
numbers.reverse();
console.log(numbers); // [ 5, 4, 3, 2, 1 ]

numbers = [1, 2, 3, 4, 5];
// numbers.sort((num1, num2) => num2 - num1);
console.log(numbers); // [ 5, 4, 3, 2, 1 ]
````

#### Solution 1
Solution using the `...` spread operator.

````javascript
let numbersCopy1 = [...numbers].sort((num1, num2) => num2 - num1);
console.log(numbersCopy1)
console.log(numbers)
````

#### Solution 2
Using a for loop

````javascript
let numbersCopy2 = []
for (let idx = numbers.length-1; idx >= 0; idx--) {
    numbersCopy2.push(numbers[idx])
}
console.log(numbersCopy2)
console.log(numbers)
````

#### Solution 3
Using `Array.prototype.slice()`

````javascript
let numbersCopy3 = numbers.slice().reverse()
console.log(numbersCopy3)
console.log(numbers)
````

#### Solution 4 (Bonus Question)
Using the forEach loop

````javascript
let reversedArray = [];

numbers.forEach((number) => {
  reversedArray.unshift(number);
});

console.log(reversedArray); // [5, 4, 3, 2, 1]
console.log(numbers);       // [1, 2, 3, 4, 5]
````

## Question 6
Suppose we build an array like this:

````javascript
let flintstones = ["Fred", "Wilma"];
flintstones.push(["Barney", "Betty"]);
flintstones.push(["Bambam", "Pebbles"]);
````

This code will create a nested array that looks like this:
```js
["Fred", "Wilma", ["Barney", "Betty"], ["Bambam", "Pebbles"]];
```

Create a new array that contains all of the above values, 
but in an un-nested format like this:
```js
[ 'Fred', 'Wilma', 'Barney', 'Betty', 'Bambam', 'Pebbles' ]
```

#### Solution 1
Using `Array.prototype.concat()` and the Spread operator 

````javascript
flintstones = [].concat(...flintstones);
````

#### Solution 2
Using `Array.prototype.reduce()`

````javascript
let array = ["Fred", "Wilma", ["Barney", "Betty"], ["Bambam", "Pebbles"]];

console.log(array)

let newArray = array.reduce((accum, element, _) => {
  return accum.concat(element)
},[])

console.log(newArray);
````

#### Solution 3
Using `Array.prototype.forEach`.

````javascript
let newFlintstones = [];
flintstones.forEach(element => {
  newFlintstones = newFlintstones.concat(element);
});
````

#### Solution 4
Using `Array.proototype.flat()`. However, this method is still in an experimental phase. 

The `flat()` method creates a new array with all sub-array elements concatenated into it recursively up to the specified depth.

If the `flintstones` array had a second level of nested arrays for example:
```js
["Fred", "Wilma", [["Barney", "Betty"]], ["Bambam", "Pebbles"]];
```

we would need to use `.flat()` specifiying a second level of 
recursion. 

```js
flintstones.flat(2)
```

````javascript
let newFlintstones2 = flintstones.flat()
console.log(newFlintstones2);
````

## Question 7
Consider the following object:

````javascript
let flintstones = { 
    Fred: 0, 
    Wilma: 1, 
    Barney: 2, 
    Betty: 3, 
    Bambam: 4, 
    Pebbles: 5 
};
````

Create an array from this object that contains only two 
elements: Barney's name and Barney's number:

```js
[ 'Barney', 2 ]
```

#### Solution 1
Using `Object.prototype.entries` and `Array.prototype.entries`.

````javascript
let flintstonesEntries = Object.entries(flintstones)
console.log(flintstonesEntries)

let barney = flintstonesEntries.find((entry) => {
    if (entry[0] === 'Barney') {
        return entry
    }
}) 
console.log(barney);
````

#### Solution 2 (Launch School Solution)

````javascript
Object.entries(flintstones)
      .filter(pair => pair[0] === "Barney")
      .shift();
````

## Question 9
Back in the stone age (before CSS), we used spaces to align things on the screen. If we have a 40-character wide table of Flintstone family members, how can we center the following title above the table with spaces?

````javascript
let title = "Flintstone Family Members";

let padding = Math.floor((40 - title.length) / 2);
// console.log(padding + title.length);

title = title.padStart(padding + title.length);
console.log(title);
````
