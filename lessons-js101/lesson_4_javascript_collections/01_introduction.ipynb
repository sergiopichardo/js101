# Introduction 

### How dow we store data? 
- The answer for most programming languages is to put them 
into structured contains, or collections. Collections come in many
different forms, and it's useful to think as lists of data. 
- 3 of the most used collections in JavaScript include `String`, `Array`, and `Object`.
- String values aren't generally considered collections since they're a primitive type, but they have some properties of collections. 
- E.g. You can extract a character or a substring from a larger string.
